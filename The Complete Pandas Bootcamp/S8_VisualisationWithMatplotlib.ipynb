## The plot method



```python
# Premable
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
titanic.head()
```

```python
titanic.info() #similar to str(titianic) in R
```

Now we import the [matplotlib](https://matplotlib.org/) and explore it.

```python
import matplotlib.pyplot as plt
```

<code>titanic.plot()</code> allows  <code>matplotlib.pyplot</code> to plot each of the columns in the dataset. Everytime we use <code>matplotlib.pyplot</code> to plot we need to call the show method (<code>plt.show()</code>).

```python
titanic.plot() # "Shift + Tab " to see more parameters
plt.show()
```

So by default the plot method creates here a line plot. On the x axis we have the index.  We have the range index from 0 to 891. These are the passenger numbers. On the y axis we have the specific values, our numerical columns. For each column here, we have a line plot.

We can see that it is difficult to make of what we have here, so we move on to create subplots for each column.

```python
titanic.plot(subplots=True) # separate subplots
plt.show()
```

As we can see the size here is small so we change the size of each graph.

```python
titanic.plot(subplots=True,figsize=(15,12)) # figsize=(height , width)
plt.show()
```

We set the <code>figsize</code> parameter to  a height of 15 and a width of 12 so that it is easier to analyse. <code>figsize=(height , width)</code>

We can create a separate x axis for each subplot. We do this by changong sharex parameter that is automatically set to True.

```python
titanic.plot(subplots=True,figsize=(15,12),sharex=False) # figsize=(height , width)
plt.show()
```

Now we can see at the bottom of each parameter, each column subplot with the range index from 0 to 891.

We can now interpret the above seeing the column of passengers survived oscillating between - and 1 as the two possible states of each passenger are survived (1) or died (0). We can continue interpreting the line plots deducing insights on the titanic dataset.

Similarly we can make the y axis have the same plot with the same scale by setting  parameter <code>sharey = True</code>.

```python
titanic.plot(subplots=True,figsize=(15,12),sharex=False, sharey=True)
plt.show()
```

Now we show how to use the line plot only on one column. Say we take the age column for example (<code>titanic.age.plot</code>). Note since we wold be using only one column the <code>subplots</code> argument would be unecessary and similarly the <code>sharex</code> <code>sharey</code>.

```python
titanic.age.plot(figsize=(10,5))
plt.show()
```

## Customization of plots

```python
titanic.age.plot(figsize=(8,5))
# plt.title("Titanic Ages", fontsize=15)
# plt.legend(loc="best", fontsize=15)
# plt.xlabel("Passenger No", fontsize=15)
# plt.ylabel("Age", fontsize=15)
# plt.grid()
plt.show()
```

```python
help(plt.plot)
```

Look at the help file on plot method in order to get specifications.
e.g. linestyle or ls: {'-', '--', '-.', ':', '', (offset, on-off-seq), ...}

```python
titanic.age.plot(figsize=(8,5), fontsize=15, c="r", ls = '--')
plt.title("Titanic Ages", fontsize=15)
# plt.legend(loc="best", fontsize=15)
# plt.xlabel("Passenger No", fontsize=15)
# plt.ylabel("Age", fontsize=15)
# plt.grid()
plt.show()
```

```python
titanic.age.plot(figsize=(8,5))
plt.title("Titanic Ages", fontsize=15) # Add a title
plt.legend(loc="best", fontsize=15) # Add a legend
plt.xlabel("Passenger No", fontsize=15) # Add a x-axis label
plt.ylabel("Age", fontsize=15) # Add  a y-axis label
# plt.grid() 
plt.show()
```

Add a grid on the background on the back.

```python
titanic.age.plot(figsize=(8,5))
plt.title("Titanic Ages", fontsize=15)
plt.legend(loc="best", fontsize=15)
plt.xlabel("Passenger No", fontsize=15)
plt.ylabel("Age", fontsize=15)
plt.grid() # Add a grid
plt.show()
```

Investigate the line plots attributes.

```python
titanic.age.plot(figsize=(8,5), 
                 fontsize=10, 
                 c="green", 
                 ls='--',
                 xlim=(200,400),
                 ylim=(20,40) )

plt.title("Titanic Ages", fontsize=15) # Add a title
plt.legend(loc="best", fontsize=15) # Add a legend
plt.xlabel("Passenger No", fontsize=15) # Add a x-axis label
plt.ylabel("Age", fontsize=15) # Add  a y-axis label
# plt.grid() 
plt.show()
```

We can also pass it a sequence/lists for the x-ticks and the y-ticks as below:

- How to create the ticks

```python
xticks = [x for x in range(0,901,50)] # from 0 to 901 by 50
xticks
```

- Create the xticks and yticks as sequences (lists) in order to pass to the plot.

```python
xticks = [x for x in range(0,901,50)] # from 0 to 901 by 50
yticks = [y for y in range(0,81,5)] # from 0 to 81 by 5
```

```python
titanic.age.plot(figsize=(8,5), 
                 fontsize=10, 
                 c="gold",
                xticks=xticks,
                yticks=yticks)
plt.title("Titanic Ages", fontsize=15) # Add a title
plt.legend(loc="best", fontsize=15) # Add a legend
plt.xlabel("Passenger No", fontsize=15) # Add a x-axis label
plt.ylabel("Age", fontsize=15) # Add  a y-axis label
# plt.grid() 
plt.show()
```

We can also rotate our x - axes by 90$^{o}$ / 45$^{o}$ .

```python
titanic.age.plot(figsize=(8,5), fontsize=10, c="gold",xticks=xticks,yticks=yticks,
                rot= 90)
plt.title("Titanic Ages", fontsize=15) # Add a title
plt.legend(loc="best", fontsize=15) # Add a legend
plt.xlabel("Passenger No", fontsize=15) # Add a x-axis label
plt.ylabel("Age", fontsize=15) # Add  a y-axis label
# plt.grid() 
plt.show()
```

Example: Two line plots together:

```python
titanic.survived.plot(figsize=(8,5)) # Age column
titanic.pclass.plot(figsize=(8,5)) # class column
plt.title("Titanic ", fontsize=15) #
plt.legend(loc='upper right', fontsize=15) # Add a legend : use 'loc' parameter to change the location
plt.xlabel("Passenger No", fontsize=15) # Add a x-axis label
plt.ylabel("Survived and pclass", fontsize=15) # Add  a y-axis label
plt.show()
```

## Histograms

```python
# Premable
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

Lets make a frequency table of each of the values:

```python
titanic.age.value_counts() # frequency table of age variable with most frequent value first and so on.
```

We can also organize our values in bins <code>bins=</code>.

```python
titanic.age.value_counts(bins=10)
```

```python
titanic.age.value_counts(bins=10)
```

We can also do this graphically where instead of passing lines as below, we pass hist.

```python
titanic.age.plot(kind="line",
                 figsize=(10,5),
                 fontsize=15)

plt.title("Titanic Ages") # Lets add a title
plt.show()
```

```python
titanic.age.plot(kind="hist",
                 figsize=(10,5),
                 fontsize=15)
plt.title("Titanic Ages Histogram") # Lets add a title
plt.show()
```

```python
titanic.age.value_counts(bins=10)
```

We can see that the histogram agrees with the data as the most frequently occuring bin is $(16.336, 24.294]$

By default we have 10 bins. The **disadvantage of using the matplotlib method is that the parameter bins is not explicitly specified in the helpfile**.

However we can specify the parameter bins and it changes the bins.

```python
titanic.age.plot(kind="hist",
                 figsize=(10,5),
                 fontsize=15,
                bins= 40)
plt.title("Titanic Ages Histogram") # Lets add a title
plt.show()
```

So we have seen that the <code>matplotlib.pyplot.plot</code> (using the pandas method) is not the optimal method for creating a histogram. Next we present better methods for creating histograms.
