# Getting started

Let's look at the titanic data set:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.float_format = '{: .3f}'.format
```

Must ensure that the titanic dataset is in the folder. [Download here](https://gist.github.com/michhar/2dfd2de0d4f8727f873422c5d959fff5) 

We then move on to import the titanic data set and start by looking at the head. Followed by tables of survival rates and lastly fitting logistic regression on the groups individually.

```python
titanic = pd.read_csv("titanic.csv")
titanic.head()
```

```python
titanic.groupby(["sex","pclass"]).survived.mean().unstack()
```

Logistic Regression:

```python
sns.lmplot(data = titanic, x="age", y="survived", aspect=1, height=8,
           col="pclass",
           logistic=True)
plt.show()
```

Logistic Regression:

```python
sns.lmplot(data = titanic, x="age", y="survived", aspect=1, height=8,
           col="sex",
           logistic=True)
plt.show()
```
